```python
import pandas as pd
from pymongo import MongoClient
```

```python
netflix = pd.read_csv(r"Desktop\Netflix_cleaned.csv")
```

```python
netflix.head()
```

```python
c = MongoClient("mongodb://localhost:27017/")
```

```python
db = c["Pythondb"]
```

```python
col = db["Netflix"]
```

```python
col.insert_many(netflix.to_dict("records"))
```

```python
print("First 5 documents in the collection:")
for doc in col.find().limit(5):    
    print(doc)
```

```python
for doc in col.find({"country": "United States", "type": "Movie"}):    
    print(doc)
```

```python
pipeline = [{"$group": {"_id": 
"$rating", "count": {"$sum": 1}}}, {"$sort": {"count": -1}}, {"$limit": 5}]
for doc in col.aggregate(pipeline):    
    print(doc)
```

```python
col.update_many({"rating": "TV-MA"},{"$set": {"rating": "Mature"}})
```

```python
col.delete_many({"release_year": {"$lt": 2000}})
```

```python
pipeline = [{"$group": {"_id": "$country", "count": {"$sum": 1}}}, {"$sort": {"count": -1}},    {"$limit": 3}]
for doc in col.aggregate(pipeline):     
    print(doc)
```
